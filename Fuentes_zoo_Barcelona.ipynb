# Listado de todas las fuentes para beber que se encuentren en el Zoológico de Barcelona
 

1. Importamos las librerías:

```python
import pandas as pd
import requests
import json 
```

2. Utilizamos la API para acceder a los datos que queremos utilizar:

```python
url='https://opendata-ajuntament.barcelona.cat/data/api/action/datastore_search'

r = requests.get(url+"?"+"resource_id=32c82e7b-2471-4576-9941-b5044312e49f")
```

3. Imprimimos las cabeceras para ver qué columnas contiene la base de datos:

```python
print(r.headers)
```

4. Asignamos a un objeto los datos:


```python
fuentes = r.json()
print(json.dumps(fuentes, indent=4))
```
