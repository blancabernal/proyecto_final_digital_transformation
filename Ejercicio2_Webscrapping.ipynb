Importar librerías

```
import requests
from bs4 import BeautifulSoup
```

Cargar url

```
url = 'https://es.wikipedia.org/wiki/Madrid'
request = requests.get(url)
contenido = request.text
```

Parsear página

```
# parseando contenido
pagina = BeautifulSoup(contenido, 'lxml')
print(pagina)
```

Acceso a secciones del HTML

```
# acceso a distintas partes de la web
print(pagina.title) # acceso solo a la etiqueta titulo
print("----------")
print(pagina.head) # acceso a toda la etiqueta head
```

Seleccionar enlaces wiki

```
import re
```

```
tablas = pagina.body.table # extrae todas las tablas de la pagina
expresion_regular = re.compile("/wiki/[^\"]+") # busca cadenas que empiecen por /wiki y que tengan cualquier caracter, excepto "
paginas_wiki = expresion_regular.findall(str(tablas))
for pagina_wiki in paginas_wiki:
  print("\t-Pagina encontrada: "+pagina_wiki)
```
